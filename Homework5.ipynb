```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Problem 1
x = np.linspace(0, 10, 100) #create x values
y = np.sin(x) #get y values
plt.plot(x, y, '--', color = 'b') #plot x against y with dashed line and blue color
plt.title('Sin(X) vs X')
plt.xlabel('X')
plt.ylabel('Sin(X)') #add labels and title
```

```python
#Problem 2
df = sns.load_dataset('tips') #get tips data as df
g = sns.histplot(df, x = 'total_bill', kde = True, bins = 20) #create histogram with kde and 20 bins
g.set_xlabel('Bill Total ($)')
g.set_title('Distribution of Bill Totals') #add title and labels
```

```python
#Problem 3
fig, ax = plt.subplots(2, 1, figsize = (6, 7)) #create subplots
x1 = df[df['time'] == 'Dinner']['tip'] #get dinner tips as x1
x2 = df[df['time'] == 'Lunch']['tip'] #get lunch tips as x2
ax[0].violinplot([x2, x1]) #create violin plot of lunch and dinner tips
ax[0].set_ylabel('Tip ($)')
ax[0].set_xlabel('Time') #add labels
ax[0].set_xticks([1, 2], ['Lunch', 'Dinner']) #add x ticks of lunch and dinner
ax[1].boxplot([x2, x1]) #create box plot of lunch and dinner tips
ax[1].set_ylabel('Tip ($)')
ax[1].set_xlabel('Time') #add labels
ax[1].set_xticks([1, 2], ['Lunch', 'Dinner']) #add x ticks of lunch and dinner
ax[0].set_title('Distribution of Tips by Time of Day') #set title
```

```python
#Problem 4
g = sns.lmplot(df, x = 'total_bill', y = 'tip', hue = 'smoker') #create scatter plot of tip vs total with linear regression line with smoker as color
g.set_axis_labels('Bill Total ($)', 'Tip ($)')
g.fig.suptitle('Tips by Smoker Status') #add title and labels
```

```python
#Problem 5
matrix = df[['total_bill', 'tip', 'size']].corr() #create correlation matrix
g = sns.heatmap(matrix, annot = True, cmap = 'vlag') #plot matrix displaying correlation coefficients
g.set_title('Correlation Matrix of Tips Dataset') #add title
```

```python
#Problem 6
g = sns.pairplot(df, hue = 'time') #make plots comparing all numerical variables
g.map_diag(sns.kdeplot) #add kde along diagonal plots
```
